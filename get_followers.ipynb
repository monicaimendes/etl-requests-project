```python
import requests
```

```python
access_token = ''
headers = {'Authorization': 'Bearer ' + access_token, 'X-GitHub-Api-Version': '2022-11-28'}
```

```python
user = 'amzn'
url = f"https://api.github.com/users/{user}/followers"
response = requests.get(url, headers=headers)
followers = response.json()
```

```python
len(followers)
```

```python
followers_list = []

page = 1

while True:
    response = requests.get(f"{url}?page={page}", headers=headers)
    followers = response.json()

    if len(followers) == 0:
        break

    followers_list.extend(followers)
    page += 1
```

```python
len(followers_list)
```

<h3>Transformando os dados</h3>

```python
followers_list[0]['login']
```

```python
followers_name_list = []

for follower in followers_list:
    followers_name_list.append(follower['login'])
```

```python
followers_name_list
```

```python
import pandas as pd
```

```python
followers_data = pd.DataFrame()
followers_data['names'] = followers_name_list
```

```python
followers_data
```

```python
followers_data.to_csv('amazon_followers.csv')
```
